```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
avocado = pd.read_csv('avocado.csv')
avocado
```

```python
region = avocado.groupby("region")["AveragePrice"].mean().head(10)
region
```

```python
region.plot(kind="bar", title="10 regions with the highest average price")
plt.ylabel("Average price")
```
